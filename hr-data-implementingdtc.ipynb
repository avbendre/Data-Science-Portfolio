### Creating a Simple mode
- Without Any Hyper-parameter Tuning
- Without Any K-Fold Cross Validation
- With the simple one-hold-out Cross Validation (train-test split)

```python
features = ['satisfaction_level', 'last_evaluation', 'number_project',
       'average_montly_hours', 'time_spend_company', 'Work_accident',
       'promotion_last_5years', 'salary', 'department_RandD',
       'department_accounting', 'department_hr', 'department_management',
       'department_marketing', 'department_product_mng', 'department_sales',
       'department_support', 'department_technical']
```

```python
# Dividing the data into train and test
```

Let us now divide it in train and test split.
We are using a basic version of train test split and we are not doing a stratified sampling as the sample is small in size.
This is for an example as well.

```python
from sklearn.model_selection import train_test_split
```

After importing the train test split, the chunk of test and train is decided, which generally is in the range of 30 to 10 percent depending on how big the data is. So, if the data is large enough, then even 10 % of the test split is sufficient to test the model behaviour. 
Random state is given as an identifier to the random test and train split generated to be same even if the data is opened and loaded many number of times.

```python
train_X, test_X, train_Y, test_Y = train_test_split(hr_df[features], hr_df['left'], test_size=0.3, random_state = 123)
```

### Creating a One-Hold Out, K-Fold Cross Validation and  One-Hold Out+ K-Fold Cross Validation model with providing hyper-parameters
- We will do all the things done abobe but this time with providing the hyper-parameters

##### One hold Out (train-test)

```python
train_X, test_X, train_Y, test_Y = train_test_split(hr_df[features], hr_df['left'], test_size=0.3, random_state = 123) 
```

```python
# Step 1- Initialize the model with the Hyperoparameters
DTC = DecisionTreeClassifier(criterion = 'entropy', max_depth = 10, min_samples_split = 5, min_samples_leaf = 2)
```

```python
# Step 2- Fit the model onto the data (during one-hold out cross validation, you fit the data on the train data)
DTC.fit(train_X, train_Y)
```

```python
train_predicted_probabilities = pd.DataFrame(DTC.predict_proba(train_X))[1]
roc_auc_score(train_Y, train_predicted_probabilities)
```

```python
test_predicted_probabilities = pd.DataFrame(DTC.predict_proba(test_X))[1]
roc_auc_score(test_Y, test_predicted_probabilities)
```

##### One hold Out + K-Fold CV

```python
results_train_k10 = cross_val_score(DTC, train_X, train_Y, cv = kfold_S, scoring='roc_auc')
results_train_k10
```

```python
np.mean(results_train_k10)
```

### Tuning Hyper-parameters with K-Fold Cross Validation

```python
# Import the GridSearchCV module

from sklearn.model_selection import GridSearchCV
```

```python
# Step 1: create a dictionary with the keys as the parameters and the values having all the possible values of the hyper-params
```

```python
params = {'criterion': ['gini','entropy'],
         'max_depth': [5,6,7],
         'min_samples_split': [2,3,4],
         'min_samples_leaf':  [3,4,5],
         'max_features': ["auto", "sqrt"]}
```

```python
# Inititalize the model but this time with hyper-parameters and K-Fold cv
DTC_H = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv=5, scoring = 'roc_auc')
```

```python
#  Fit the Model
DTC_H.fit(train_X, train_Y)  # Can do it on the complete data. Technically it will be fine bcz we are using KFoldCV. But to be extra cautious
                             # we will use train data here and then will apply the model onto the left out test data
                             # Hence a combination of K-Fold and One Hold Out
```

```python
DTC_H.best_params_
```

```python
DTC_H.best_score_
```

```python
# Using the above found hyper-params to come up with a model
```

```python
DTC_F = DecisionTreeClassifier(criterion = 'entropy', max_depth = 7, min_samples_split = 3, min_samples_leaf = 4, max_features = 'auto')
```

```python
DTC_F.fit(train_X, train_Y)
```

```python
train_predicted_probabilities = pd.DataFrame(DTC_F.predict_proba(train_X))[1]
roc_auc_score(train_Y, train_predicted_probabilities)
```

```python
test_predicted_probabilities = pd.DataFrame(DTC_F.predict_proba(test_X))[1]
roc_auc_score(test_Y, test_predicted_probabilities)
```
